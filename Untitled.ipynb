```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import charset_normalizer
with open("Airline Dataset Updated.csv", 'rb') as rawdata:
    result = charset_normalizer.detect(rawdata.read(10000))

# check what the character encoding might be
print(result)
```

```python
df=pd.read_csv("Airline Dataset Updated.csv",encoding="utf-8")
print(df.shape)
df.head(10)
```

```python
#Cleaning data
column_names=['Passenger ID', 'First Name', 'Last Name','Airport Country Code','Airport Continent','Pilot Name']
df=df.drop(column_names,axis=1)
```

```python
df.dtypes
```

```python
df["Nationality"].nunique()
```

```python
nation_count = df['Nationality'].value_counts().rename_axis('Nationality').reset_index(name="Count")
nation_count
```

```python
#top 10 nation wit most passengers
top_nation=nation_count.nlargest(10,"Count")
sns.barplot(y="Nationality",x="Count",data=top_nation,palette="flare")
plt.xlabel("Nationality")
plt.ylabel("Number of passengers")
plt.title("Top 10 nation with most passengers")
plt.show()
```

```python
##top 10 nation wit least passengers
least_nation=nation_count.nsmallest(10,"Count")
sns.barplot(y="Nationality",x="Count",data=least_nation,palette="flare")
plt.xlabel("Nationality")
plt.ylabel("Number of passengers")
plt.title("Top 10 nation with least passengers")
plt.show()
```

```python
df["Gender"].nunique()
```

```python
#which gender seems to travel most

gender_count=df["Gender"].value_counts().rename_axis('Gender').reset_index(name="Count")
gender_count
```

```python
sns.barplot(x="Gender",y="Count",data=gender_count,palette='flare')
plt.xlabel("Gender")
plt.ylabel("Number of passengers")
plt.show()
```

```python
df["Airport Name"].unique()
```

```python
airport_passengers_count=df["Airport Name"].value_counts().rename_axis("Airport Name").reset_index(name="Count")
airport_passengers_count
```

```python
#top 10 airport with most passengers
most_airport_passengers_count=airport_passengers_count.nlargest(10,"Count")
sns.barplot(y="Airport Name",x="Count",data=most_airport_passengers_count,palette="flare")
plt.ylabel("Airport Name")
plt.xlabel("Number of passengers")
plt.title("Top 10 airport wit most passengers")
plt.show()
```

```python
#On Time flight
ontime_df=df[df["Flight Status"]=="On Time"]
ontime_df
```

```python
ontime_df['Airport Name'].nunique()
```

```python
airport_performance=ontime_df["Airport Name"].value_counts().rename_axis("Airport Name").reset_index(name="Count")
airport_performance
```

```python
#top performing airport
top_performing=airport_performance.nlargest(5,"Count")
sns.barplot(y="Airport Name",x="Count",data=top_performing,palette="flare")
plt.ylabel("Airport Name")
plt.xlabel("Number of Flights")
plt.title("Top 5 Performing Airport")
plt.show()
```

```python
#Catageorizing the according to age group
Age_category=['Children','Adolescence','Adult']
conditions=[(df["Age"]<=12),(df["Age"]>12)&(df["Age"]<=18),(df["Age"]>18)]
df["Age Category"]=np.select(conditions,Age_category,default="unknown")
df
```

```python
ageCategory_count=df["Age Category"].value_counts().rename_axis("Age Category").reset_index(name="Count")
ageCategory_count
# gender_count=df["Gender"].value_counts().rename_axis('Gender').reset_index(name="Count")
```

```python
plt.figure(figsize=(8,6))
sns.barplot(y="Count",x="Age Category",data=ageCategory_count,palette='flare')
plt.xlabel("Age Category")
plt.ylabel("Number of Passengers")
plt.title("Passengers according to Age category")
plt.show()
```

```python
#Arrival airport
df['Arrival Airport'].replace('0', 'Unknown', inplace=True)
arrivalAirport=df["Arrival Airport"].value_counts().rename_axis("Arrival Airport").reset_index(name="Count")
arrivalAirport
```

```python
arrivalAirport_count=arrivalAirport.nlargest(10,"Count")
```

```python
plt.figure(figsize=(8,6))
sns.barplot(y="Arrival Airport",x="Count",data=arrivalAirport_count,palette='flare')
plt.xlabel("Number of Passengers")
plt.ylabel("Arrival Airport")
plt.show()
```
